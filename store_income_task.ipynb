```python
# Load up store_income_data.csv
import pandas as pd
import datetime
from fuzzywuzzy import process

# Load data
data = pd.read_csv('cleaned_store_income.csv')
```

 1. Take a look at all the unique values in the "country" column. Then, convert the column to lowercase and remove any trailing white spaces.

```python
# View all unique values in the "country" column
print(data['country'].unique())

# Convert the "country" column to lowercase and remove trailing white spaces
data['country'] = data['country'].str.lower().str.strip()
```

2. Note that there should only be three separate countries. Eliminate all variations, so that 'South Africa', 'United Kingdom' and 'United States' are the only three countries.

```python
countries_needed = ['south africa', 'united kingdom', 'united states']

# Fuzzy match and replace function
def replace_matches_in_column(df, column, string_to_match, min_ratio=80):
    strings = df[column].unique()
    
    # Find close matches
    matches = process.extract(string_to_match, strings, limit=100)
    close_matches = [match for match, score in matches if score >= min_ratio]
    
    # Replace matches with the string_to_match
    rows_with_matches = df[column].isin(close_matches)
    df.loc[rows_with_matches, column] = string_to_match
    
for country in countries_needed:
    replace_matches_in_column(data, 'country', country)
```

3. Create a new column called days_ago in the DataFrame that is a copy of the 'date_measured' column but instead it is a number that shows how many days ago it was measured from the current date. Note that the current date can be obtained using datetime.date.today().

```python
from datetime import date

# Convert 'date_measured' to datetime format
data['date_measured'] = pd.to_datetime(data['date_measured'], errors='coerce', dayfirst=True)

# Get today's date
today = datetime.date.today()

# Create 'days_ago' column
data['days_ago'] = data['date_measured'].dt.date.apply(lambda x: (today - x).days if pd.notnull(x) else None)

# View the result
print(data[['date_measured', 'days_ago']].head())

# Save cleanes file
data.to_csv('cleaned_store_income.csv', index=False)
```
